# Analyzing Data with Pandas and Visualizing Results with Matplotlib
Assignment Submission

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import numpy as np

# Enable seaborn styling
sns.set()
```

## Task 1: Load and Explore the Dataset

```python
try:
    # Load Iris dataset from sklearn
    iris = load_iris(as_frame=True)
    df = iris.frame  # Dataset as a pandas DataFrame

    # Display the first few rows
    print("First 5 rows of the dataset:")
    print(df.head())

    # Explore the structure
    print("\nData types and null values:")
    print(df.info())

    print("\nChecking for missing values:")
    print(df.isnull().sum())

except Exception as e:
    print(f"Error loading dataset: {e}")
```

## Task 2: Basic Data Analysis

```python
# Compute basic statistics
print("\nBasic Statistics:")
print(df.describe())

# Perform grouping by 'target' (species)
grouped = df.groupby('target').mean()
print("\nMean values grouped by Species (target):")
print(grouped)

# Map target numbers to species names
species_map = dict(zip(range(3), iris.target_names))
df['species'] = df['target'].map(species_map)

# Check for patterns
print("\nAverage measurements by species:")
print(df.groupby('species').mean())
```

## Task 3: Data Visualization

```python
# Line chart - Sepal length over sample index
plt.figure(figsize=(10,6))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length')
plt.title('Sepal Length over Sample Index')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar chart - Average petal length per species
plt.figure(figsize=(8,5))
sns.barplot(x='species', y='petal length (cm)', data=df, ci=None)
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram - Distribution of petal width
plt.figure(figsize=(8,5))
plt.hist(df['petal width (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot - Sepal length vs. Petal length
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

## Findings and Observations:
- Setosa species have smaller petal dimensions compared to Versicolor and Virginica.
- Petal width distribution indicates different ranges for each species.
- Positive correlation exists between Sepal Length and Petal Length.
